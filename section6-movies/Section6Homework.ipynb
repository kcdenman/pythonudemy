```python
biz = pd.read_csv('moviesimbd.csv', encoding = 'latin1')
```

```python
#Clean up column headers

biz.columns = ['DayOfWeek', 'Director', 'Genre', 'MovieTitle', 'ReleaseDate',
       'Studio', 'AdjustedGross_M', 'Budget_M', 'Gross_M',
       'IMDbRating', 'MovieLensRating', 'Overseas_M', 'Overseas_Pct',
       'Profit_M', 'Profit_Pct', 'Runtime_min', 'US_M',
       'Gross_Pct_US']
```

```python
#import data science libraries

from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
#Filter out genres of interest

filter_genre = (biz.Genre == 'action')\
                    | (biz.Genre == 'comedy')\
                    | (biz.Genre == 'adventure')\
                    | (biz.Genre == 'animation')\
                    | (biz.Genre == 'drama')

filter_studio = (biz.Studio == 'Buena Vista Studios')\
                    | (biz.Studio == 'Sony')\
                    | (biz.Studio == 'Universal')\
                    | (biz.Studio == 'WB')\
                    | (biz.Studio == 'Paramount Pictures')\
                    | (biz.Studio == 'Fox')
```

```python
subset = pd.DataFrame(biz[filter_genre & filter_studio])
```

```python
subset.describe()
```

```python
#check on filters
for gen in subset.Studio.unique():
    print(gen)
```

```python
figure, axes = plt.subplots(figsize=(15,10))

v1 = sns.swarmplot(data=subset,x='Genre',y='Gross_Pct_US',hue='Studio')
v2 = sns.boxplot(data=subset,x='Genre', y='Gross_Pct_US')

#-----FORMATTING-----#
v1.set_title("Domestic Gross % by Genre", fontsize=24, \
         color='Black', fontname='Roboto')
v1.set_ylabel("Gross % US", fontsize=18, color='Black', fontname='Roboto')
v1.set_xlabel("Genre", fontsize=18, color='Black', fontname='Roboto')
# m2.tick_params(axis='both',**{'labelsize':18,'pad':10})

plt.legend(loc=(1,0))
plt.show()
```
