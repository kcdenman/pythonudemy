```python
###Keyword Arguments in Python (advanced tutorial)
```

```python
#market size:
# a keyword argument is an argument passed through the function
# using a dictionary of keywords.
# pass a dictionary sccater_kws from matplotlibs.pyplot 

vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', \
                  fit_reg=False, hue='IncomeGroup', size=10, \
                 scatter_kws={"s":200}) 
```
